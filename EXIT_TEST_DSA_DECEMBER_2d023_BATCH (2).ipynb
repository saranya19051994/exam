# filling missing value

```python
data['Work-Life Balance']= data ['Work-Life Balance'].fillna(data['Work-Life Balance'].mode()[0]) # Use mode for categorical data
data['Performance Rating'] = data['Performance Rating'].fillna(data['Performance Rating'].mode()[0])
data['Number of Promotions']= data ['Number of Promotions'].fillna(data['Number of Promotions'].mode()[0])
data['Education Level']= data ['Education Level'].fillna(data['Education Level'].mode()[0])
data['Marital Status']= data ['Marital Status'].fillna(data['Marital Status'].mode()[0])
data['Number of Dependents'] = data['Number of Dependents'].fillna(data['Number of Dependents'].mode()[0])
data['Job Level'] = data['Job Level'].fillna(data['Job Level'].mode()[0])
data['Company Size'] = data['Company Size'].fillna(data['Company Size'].mode()[0])
data['Company Tenure'] = data['Company Tenure'].fillna(data['Company Tenure'].mode()[0])
data['Remote Work'] = data['Remote Work'].fillna(data['Remote Work'].mode()[0])
data['Leadership Opportunities'] = data['Leadership Opportunities'].fillna(data['Leadership Opportunities'].mode()[0])
data['Innovation Opportunities'] = data['Innovation Opportunities'].fillna(data['Innovation Opportunities'].mode()[0])
data['Company Reputation'] = data['Company Reputation'].fillna(data['Company Reputation'].mode()[0])
data['Employee Recognition'] = data['Employee Recognition'].fillna(data['Employee Recognition'].mode()[0])
```
